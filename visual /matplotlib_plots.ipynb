# Working with multiple figures and axes

```python
def f(p):
  return np.exp(-p)*np.cos(2*np.pi*p)
p1= np.arange(0.0,5.0,0.1)
p2=np.arange(0.0,5.0,0.02)

plt.figure(1)
plt.subplot(211) # two rows and one columns and figure one(211 means)
plt.plot(p1,f(p1),'b-',p2,f(p2))

plt.subplot(212) # two rows and one columns and figure two(212 means)
plt.plot(p2,np.cos(2*np.pi*p2),'r--')

plt.figure(2)
plt.subplot(211) # two rows and one columns and figure two(212 means)
plt.plot(p2,np.cos(2*np.pi*p2),'g--')
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3,4])
plt.subplot(212)
plt.plot([1,4,9,16])

plt.figure(2)
plt.plot([1,2,3,4])

plt.figure(1)
plt.subplot(211)
plt.title("Subplot as 1,2,3,4")
plt.show()
```

# Bar Chart and Histogram

```python
g=[2,4,6,8,10]
h=[3,6,9,12,15]

g1=[1,5,3,9,4]
h1=[2,7,6,3,10]

plt.bar(g,h,label='bar1')
plt.bar(g1,h1,label='bar2',facecolor='yellow',color='red') #If both *color* and *facecolor are given, *facecolor* takes precedence
plt.legend()
plt.xlabel('g')
plt.ylabel('h')
plt.title('Experimenting with bar plot')
plt.show()
```

```python
days = [1,2,3,4,5]
sleeping = [6,7,8,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2,]
playing = [8,5,7,8,13]
```

```python
slices=[7,2,3,14]
activities=['sleeping','eating','working','playing']
cols=['c','m','r','b']
```

### Pie Plot features


- autopct enables you to display the percent value using Python string formatting

- The labels parameter must be an array

- The default start angle is at the x-axis, parameter is defined with an angle in degrees, default angle is 0

- Explode makes the wedges to stand out,Each value represents how far from the center each wedge is displayed

- Add a shadow to the pie chart by setting the shadows parameter to True

- The colors parameter, if specified, must be an array with one value for each wedge.

'r' - Red

'g' - Green

'b' - Blue

'c' - Cyan

'm' - Magenta

'y' - Yellow

'k' - Black

'w' - White

- To add a list of explanation for each wedge, use the legend() function.

```python
# autopct enables you to display the percent value using Python string formatting
# The labels parameter must be an array
# the default start angle is at the x-axis, parameter is defined with an angle in degrees, default angle is 0
#  explode makes the wedges to stand out,Each value represents how far from the center each wedge is displayed
plt.pie(slices, labels=activities,colors=cols,startangle=90,shadow=True,explode=(0,0.1,0,0),autopct='%.2f')
plt.title('Pie Plot')
plt.show()
```

# Stack Plot

```python
days = [1,2,3,4,5]
sleeping = [6,7,8,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2,]
playing = [8,5,7,8,13]
```

```python
plt.plot([],[],color='m',label='sleeping',linewidth=5)
plt.plot([],[],color='c',label='eating',linewidth=5)
plt.plot([],[],color='r',label='working',linewidth=5)
plt.plot([],[],color='k',label='playing',linewidth=5)
plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot of Activities')
plt.legend()
plt.show()
```
